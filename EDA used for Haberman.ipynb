# Mean, Variance and Std-dev

```python
#Mean, Variance, Std-deviation,  
print("Means:")
print(np.mean(df_1["Age"]))
#Mean with an outlier.
print(np.mean(np.append(df_1["Age"],50)));
print(np.mean(df_2["Age"]))

print("\nStd-dev:");
print(np.std(df_1["Age"]))
print(np.std(df_2["Age"]))
```
